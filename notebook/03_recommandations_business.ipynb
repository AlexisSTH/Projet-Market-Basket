## 6. Recommandations business

Sur la base des règles d'association identifiées, voici les actions concrètes proposées :

### 6.1 Placement en rayon (cross-merchandising)
- **Fromage emballé à proximité du pain** : lift de 1.56, ces produits sont fréquemment achetés ensemble. Placer un présentoir de fromage à côté du rayon boulangerie peut augmenter les ventes croisées.
- **Lait à proximité du pain** : lift de 1.49, association type "petit-déjeuner". Un rappel produit en bout de rayon boulangerie serait pertinent.

### 6.2 Promotions croisées (bundles)
- **Offre "plateau apéro"** : fromage emballé + pain → proposer un bundle avec réduction.
- **Offre "cuisine maison"** : herbes fraîches + légumes → un pack promo pour encourager la cuisine.

### 6.3 Recommandations en ligne (cross-sell)
- Quand un client ajoute du pain à son panier → suggérer du fromage emballé (confidence 36%) et du lait (confidence 36%).
- Quand un client ajoute des herbes fraîches → suggérer des légumes emballés (confidence 55%).

### 6.4 Limites de l'analyse
- L'échantillon couvre 10 000 clients sur 206 209 — les résultats mériteraient validation sur le dataset complet.
- Les associations au niveau rayon sont moins précises qu'au niveau produit (ex : "fromage" est vague).
- Le lift reste modeste (~1.5) — les associations sont réelles mais pas extrêmement fortes.
- Une analyse temporelle (par saison, par heure) pourrait révéler des associations contextuelles plus ciblées.
